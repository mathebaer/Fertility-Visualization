```python
print(df.columns)
feature_list = df.drop('Score', axis=1).drop('Overall rank', axis=1).drop('Country or region', axis=1).columns
print(feature_list)
```

##### get matrix of correlations

```python
correlations = pd.DataFrame(columns=['feature1', 'feature2', 'correlation', 'correlation_abs'])
counter = 0
for i in feature_list:
    for j in feature_list:
        correlations.loc[counter] = [i,j, df[i].corr(df[j]), abs(df[i].corr(df[j]))]
        counter += 1
#correlations = correlations.sort_values(by='correlation_abs', ascending=False)
```

```python
g = sns.PairGrid(df[['GDP per capita',
       'Social support', 'Healthy life expectancy',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption', 'Score']])
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
```

```python
#!pip install geopandas
import geopandas as gpd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# fix some countries
df['Country or region'] = df['Country or region'].replace('Czech Republic', 'Czechia')
df['Country or region'] = df['Country or region'].replace('United States', 'United States of America')


# Load the world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Merge the world dataframe with your dataframe
merged = world.set_index('name').join(df.set_index('Country or region'))

# Create the figure and the axes
fig, ax = plt.subplots(1, 1, figsize=(15, 8))

# Create the map.
merged.plot(column='Score', ax=ax, legend=True, cmap='YlGn', 
            legend_kwds={'shrink': 0.4, 'label': "Happiness\nScore"})

# Remove the axis
ax.axis('off')

# Add a title
plt.title('Happiness by Country')

plt.show()


```
